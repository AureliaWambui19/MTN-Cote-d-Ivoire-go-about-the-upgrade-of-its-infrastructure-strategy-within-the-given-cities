DATA PREPARATION



```
# Removing duplicates
CellsGeo.drop_duplicates( inplace = True) 
CellsGeo
#   removed 43 duplicate records
```

```
# Removing missing data
CellsGeoClean = CellsGeo.dropna()
CellsGeoClean
#   removed 68 missing records
```

```
#rename colunm for dataset 1
Telcom1.rename(columns={"PRODUTC": "PRODUCT"}, inplace=True)
#previewing first row to check column update
Telcom1.head(1)
```

```
#rename colunm for dataset 2

Telcom2.rename(columns={"DATE_TIME":"DATETIME", "DW_A_NUMBER": "DW_A_NUMBER_INT", "DW_B_NUMBER":"DW_B_NUMBER_INT"}, inplace=True)
#previewing first row to check column update
Telcom2.head(1)
```

```
#rename colunm for dataset 3
Telcom3.rename(columns={"DATE_TIME":"DATETIME", "CELLID": "CELL_ID", "SIET_ID":"SITE_ID"}, inplace=True)
#previewing first row to check column update
Telcom3.head(1)
```

```
 #concating the three datasets(TeTelcom1,Telcom2,Telcom3)
 AllTelcoms=pd.concat([Telcom1,Telcom2,Telcom3], axis= 0)
  #previewing the last 2 rows to check on concatination
 AllTelcoms.tail(2)
```

```
# Dropping unnecessary columns 
AllTelcoms = AllTelcoms.drop(columns = ['COUNTRY_A' ,	'COUNTRY_B' ,'CELL_ON_SITE'] ,axis =1)
# previewing the last two rows to check on the dropped columns
AllTelcoms.head(2)
```

```
# to split the date and time column into separate columns


# create a new column and split the date_time column into 2
AllTelcoms.head()
AllTelcoms[['DATE', 'TIME']] = AllTelcoms.DATETIME.str.split(expand=True)
# previewing the last row to check on the newly created columns
AllTelcoms.head(1)
```

```
# merging datasets
TelGeo = AllTelcoms.merge(CellsGeoClean, left_on="SITE_ID",right_on="SITE_CODE")
# previewing first two rows to check on the merge
TelGeo.head(2)
```
